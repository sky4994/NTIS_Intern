# 1. mecab을 활용한 형태소 분리

```python
import itertools
from konlpy.tag import Mecab
mecab = Mecab('c:\mecab\mecab-ko-dic')

col1 = data2['요약문_연구목표'].dropna()
col2 = data2['요약문_연구내용'].dropna()
col3 = data2['요약문_기대효과'].dropna()
col4  = data2['요약문_한글키워드'].dropna()

k1 = []

for i in col1:
    k1.append(mecab.nouns(i))
for i in col2:
    k1.append(mecab.nouns(i))
for i in col3:
    k1.append(mecab.nouns(i))
for i in col4:
    k1.append(mecab.nouns(i))

nouns = list(itertools.chain.from_iterable(k1))

```

```python
from collections import Counter

count_nouns = Counter(nouns)
print(count_nouns)


count_nouns = dict(count_nouns)
nouns_list = list(count_nouns.keys())
count_list = list(count_nouns.values())

df = pd.DataFrame({'nouns' : nouns_list,
                   'counts' : count_list})
df.head()

df.sort_values('counts', ascending = False)
df.sort_values('nouns', ascending = True)
```

```python
import re
nouns_str = ','.join(nouns)

# data4 = re.sub('[ㄱ-ㅎ \u3131-\u3163\uac00-\ud7a3]+', '', nouns_str)
# rule = re.compile('[^ ㄱ-ㅣ가-힣]+')
data4 = re.sub('[ㄱ-ㅎ ㅏ-ㅣ]+',',', nouns_str)

rule = re.compile('[^ ㄱ-ㅣ가-힣]+')
data4 = rule.sub(',', data4)
nouns1 = data4.split(',')
nouns1

from collections import Counter

count_nouns = Counter(nouns1)
# print(count_nouns)


count_nouns = dict(count_nouns)
nouns_list = list(count_nouns.keys())
count_list = list(count_nouns.values())

df = pd.DataFrame({'nouns' : nouns_list,
                   'counts' : count_list})
df.head()

# df.sort_values('counts', ascending = False)
df.sort_values('nouns', ascending = True)
```

```python
df.to_csv('mecab_산업부.중기청.csv', index=False, encoding = 'cp949')
```

# 2. konlpy-okt( =twitter)로 형태소 분리

```python
import warnings
warnings.simplefilter("ignore")

import itertools
from konlpy.tag import Okt
okt = Okt()

col1 = data2['요약문_연구목표'].dropna()
col2 = data2['요약문_연구내용'].dropna()
col3 = data2['요약문_기대효과'].dropna()
col4  = data2['요약문_한글키워드'].dropna()

k1 = []

for i in col1:
    k1.append(okt.nouns(i))
for i in col2:
    k1.append(okt.nouns(i))
for i in col3:
    k1.append(okt.nouns(i))
for i in col4:
    k1.append(okt.nouns(i))

okt_nouns = list(itertools.chain.from_iterable(k1))

okt_nouns
```

```python
from collections import Counter

count_nouns = Counter(okt_nouns)
# print(count_nouns)


count_nouns = dict(count_nouns)
nouns_list = list(count_nouns.keys())
count_list = list(count_nouns.values())

df2 = pd.DataFrame({'nouns' : nouns_list,
                   'counts' : count_list})
df2.head()

df2.sort_values('counts', ascending = False)
df2.sort_values('nouns', ascending = True)
```

```python
df2.to_csv('okt_산업부.중기청.csv', index=False, encoding = 'cp949')
```
